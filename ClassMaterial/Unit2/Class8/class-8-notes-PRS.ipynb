# Getting information from time

## Pandas & Time - Jonathan's language

* Horizontal data extraction
    * Looking at internal attributes of a particular date
    * e.g. using the dt attribute, it doesn't look at a period

* Directional data extraction - looks at:
    * Rolling averages
    * Momentum
    * Comparisons over time
    
    * Good for:
        * Comparing rates of growth
        * Demand forecasting
        * Financial analysis


## Useful methods for directional data
* shift()
* diff()
* pct_change()

```python
# Need to put in parse_dates as a list!!!

sales = pd.read_excel('../data/superstore.xls', parse_dates=['Order Date', 'Ship Date'])
```

```python
sales.head()
```

```python
sales.info()
```

```python
sales['Profit']
```

```python
sales = sales.sort_values(by='Order Date')
```

```python
sales.head(10)
```

```python
profits = sales.groupby('Order Date')[['Profit']].sum()

# This gives daily profits
```

```python
profits
```

```python
profits.shift()
# Advances each row by one / pushes it down (i.e. increase its position by one)
```

```python
# statistical models evaluate each row as a sample
# so often need to create shifted data
```

```python
profits['yesterday'] = profits.shift()
```

```python
profits
```

```python
profits['Profit'].shift(2)
# thist shifts by 2 time setps
```

```python
profits['profits2daysago'] = profits['Profit'].shift(2)
```

```python
profits.head()
```

```python
profits['Profit'].pct_change()
# Calculates change vs prior row - takes an argument for number of periods
```

```python
profits['1dayprofitchange'] = profits['Profit'].pct_change()
profits['2dayprofitchange'] = profits['Profit'].pct_change(2)
```

```python
profits.head(15)
```

## Moving averages

```python
profits['Profit'].rolling(7).mean().head(20)

# Rolling creates a window of data, that the function is applied to
```

```python
profits['Profit'].head(15)
```

```python
profits['Profit'][:7].mean()
```

```python
profits['Profit'][1:8].mean()
```

```python
profits['Profit'][2:9].mean()
```

```python
profits['Profit'].rolling(7).std().head(15)
```

* Lots of time based models - they are about chasing trends, or catching when they shift

```python
profits['7daystd'] = profits['Profit'].rolling(7).std()
```

```python
profits.head(10)
```

# Hierarchical data
**Now for the twist!:**

* We have several different levels of hierarchy that we normally will want to deal with.

```python
sales.columns
```

```python
customer_profits = sales.groupby(['Customer ID','Order Date']).sum().reset_index()
```

```python
customer_profits
```

```python
customer_profits.groupby(['Customer ID','Order Date']).apply(lambda x: x['Profit'].rolling(7).mean())
```

```python
customer_profits.set_index(['Customer ID','Order Date'],inplace=True)
# This takes some of the rows, and pushs them to the index
```

```python
customer_profits.head()
```

```python
# Does the operation at level of customer id
customer_profits.groupby(level=0).sum()
```

```python
# Does the operation at level of dates
customer_profits.groupby(level=1).sum()
```

```python
customer_profits.groupby(level=0).shift()
# This shifts at the level of each individual customer
```

```python
customer_profits.groupby(level=0)[['Profit']].rolling(3).mean().head(10)
```

```python
customer_profits['Profit'].astype(str)
# this recasts the variable as a string
# e.g. in our data set might use this to stop Row ID from calculating
```
