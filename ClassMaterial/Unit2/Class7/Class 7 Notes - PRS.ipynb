### Looking at merging data
So a key common column here are the "Order ID" column between our two sheets.

```python
returns.Returned.unique()
```

```python
returns.shape
```

**So:** What do we do when we want to see if some of our items have returns in our overall data set?

```python
?pd.merge
# Note that the default is inner join   
# Can use on here since 
```

```python
sales.merge(returns, how='inner', on='Order ID')
# Here we get 800 rows back since we only show items in both data sets
```

```python
sales.merge(returns, how='left', on='Order ID')
```

```python
sales.merge(returns, how='left', on='Order ID').shape
# This time we don't loose any values, but get the extra columns
```

```python
returns.merge(sales, how='right', on='Order ID').shape
# Swapping returns / sales and shifting to a right join produces the same result
```

```python
# You can specify columns if different names
returns.merge(sales, how='right', left_on='Order ID', right_on='Order ID').shape
```

```python
# To merge on multiple columns
# You can specify columns if different names, provide an ordered list.
# This is a fairly common requirement with a lot of data sets
returns.merge(sales, how='right', left_on=['Order ID','Returned'], right_on=['Order ID','Returned']).shape
```

```python
# You can return specific columns, since this operation returns a data frame
returns.merge(sales, how='right', left_on='Order ID', right_on='Order ID').iloc[:,:4]
```

## Can write customer functions to applied to each grouping

```python
master.groupby('id')[['visitors','Days']].agg([np.sum,'mean','count'])
```

```python
def count_nulls(dataframe):
    return dataframe.isnull().sum()

# Gives the data frame as an argument
```

```python
count_nulls(master)
```

```python
master.groupby('id')[['visitors','Days']].agg([np.sum,'mean','count',count_nulls])
```

```python
# Think of each of the ids as their own little data sets, that the agg 
# then applies all the functions over on the right
```

# Lambda functions

```python
# Lambda functions are really short, throw-away functions

# idea is that it saves you from formerly defining a funtion
# tend to see them a fair bit in certain parts of pandas code

# these are nice because you can build them straight into the .apply()
# logic for the data frame to allow custom logic to be applied
```

```python
fun = lambda x: x**2 # here x represents the argument
```

```python
fun(5)
```

```python
fun(2)
```

```python
master['visitors'].max() - master['visitors'].min()
```

```python
master.groupby('id')['visitors'].apply(lambda x: x.max() - x.min())
```

```python
master.groupby('id')['visitors'].apply(lambda x: x.max() - x.min()).reset_index()
```

```python
# Another way

master.groupby('id').apply(lambda x: x['visitors'].max() - x['visitors'].min()).reset_index()
```
