Any time a statement returns true / false values in pandas. You can pass it into your data set to give you a limited set of rows.

You need to see both the **INDEX** and the **TRUE/FALSE** values.

```python
df[df['day_of_week'] == 'Wednesday'] # For example - this just returns columns that are Wednesday.
```

```python
(df['day_of_week'] == 'Wednesday') | (df['day_of_week'] == 'Thursday')
```

```python
df[(df['day_of_week'] == 'Wednesday') | (df['day_of_week'] == 'Thursday')]
```

```python
df[(df['day_of_week'] == 'Wednesday') & (df['visitors'] == 2)]
```

```python
df[df['day_of_week'] == 'Wednesday']['visitors']
```

```python
df[df['day_of_week'] == 'Wednesday'][['visitors','reserve_visitors']].mean()
```
